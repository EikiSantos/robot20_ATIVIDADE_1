# Equalização de histograma

```python
rintin = cv2.imread("RinTinTin.jpg")
```

```python
rintin.shape
```

```python
rintin_gray = cv2.cvtColor(rintin, cv2.COLOR_RGB2GRAY)
```

```python
rintin_gray.shape
```

```python
plt.imshow(rintin_gray, cmap="Greys_r", vmin=0, vmax=255)
```

```python

hist,bins = np.histogram(rintin_gray.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()

rintin_01 = rintin_gray/58

rintin_co = rintin_01*255
```

```python
img = cv2.imread("cena_canto_sala.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)

print(img.shape)
```

```python
img_r = img[:,:,0]
plt.imshow(img_r, cmap="Greys_r");
```

```python
img_b = img[:,:,2]
plt.imshow(img_b, cmap="Greys_r");
```

```python
hist,bins = np.histogram(img_r.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.hist(img_r.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()

hist,bins = np.histogram(img_b.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.hist(img_b.flatten(),256,[0,256], color = 'b')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
img_r.shape
```

```python
img_saida = img_r.copy()
```

```python
img_saida.shape
```

```python
img_saida[:] = 0
plt.imshow(img_saida, cmap="Greys_r")

img_saida = np.zeros((3024, 4032), dtype=np.uint8)
for i in range(img_saida.shape[0]):
    for j in range(img_saida.shape[1]):
        img_saida[i][j] = img_r[i][j]
```

```python
for i in range(img_saida.shape[0]):
    for j in range(img_saida.shape[1]):
        if img_r[i][j] >= 210 and img_b[i][j] < 20:
            img_saida[i][j] = 255
        else:
            img_saida[i][j] = 0
            
plt.imshow(img_saida, cmap="Greys_r")
```

Abra a imagem `hall_box_battery_atividade3.png`  e a converta para tons de cinza de `0` a `255`.

Forneça como saída uma uma cópia da imagem  em que cada pixel  recebe o módulo da subtração entre o pixel que vem depois e o que vem antes dele (na horizontal).

Obs.: O conceito de *antes* e *depois* pode ser na horizontal ou na vertical

Apresente a imagem resultado, deve ficar parecida com a imagem `gabarito_atividade_3.png`.

<img src="gabarito_atividade_3.png" width=400>

```python
image = cv2.imread("hall_box_battery_atividade3.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(image, cmap="Greys_r")
```

```python
saida = image.copy()

saida[:] = 0
for i in range(saida.shape[0]):
    for j in range(saida.shape[1]):
        saida[i][j] = image[i][j]
```

```python
image[0]
```

```python
for i in range(saida.shape[0]):
    for e in range(saida.shape[1]):
        if i==e==0 or i==image.shape[0]-1:
            saida[i][e] = image[i][e]
        else:
            saida[i][e] = abs(int(image[i+1][e]) - int(image[i-1][e]))
```

```python
plt.imshow(saida, cmap="Greys_r")
```
